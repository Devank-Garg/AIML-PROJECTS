```python
import seaborn as sns 
import matplotlib.pyplot as plt  # importing the visualisation libraries
```

```python
#checking for outliers
#plotting the boxplots helps in identifying the outliers very easily
#outliers will be shown outside of the “whiskers” of the plot, each side representing the minimum and maximum values of the dataset, respectively.
cols = ['carat','depth', 'table', 'x', 'y', 'z','price' ]
for i in cols:
    sns.boxplot(DF[i],whis=1.5) # ploting the boxplots for each column
    plt.show()
```

```python
# function for removing the outlier
def del_outlier(cols):
    sorted(cols)
    Q1,Q3=np.percentile(cols,[25,75]) # Q1= lower quaterile Q3= upper quaterile 
    IQR=Q3-Q1 #Interquaterile Range, equal to the difference between 75th and 25th percentiles i.e Q3 and Q1
    lower_range= Q1-(1.5 * IQR) # updated lower range 
    upper_range= Q3+(1.5 * IQR) # updated upper range
    return lower_range, upper_range
```

```python
for column in DF[cols].columns:
    Lower_Range,Upper_Range=del_outlier(DF[column])
    DF[column]=np.where(DF[column]>Upper_Range,Upper_Range,DF[column]) # Updating the columns by deleting  those which lie above the upper limit
    DF[column]=np.where(DF[column]<Lower_Range,Lower_Range,DF[column]) # Updating the columns by deleting  those which lie below the lower limit
```

```python
cols = ['carat','depth', 'table', 'x', 'y', 'z','price' ] 
for i in cols:
    sns.boxplot(DF[i],whis=1.5) # re-ploting the boxplots for each column after removing the outliers
    plt.show()
```

#Now let's begin the prediction part
## For prediction, we will be using the linear regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# But before starting the model training, 
#we will have to first convert the categorical features into numerical features i.e. cut, clarity and color
## Since there are more than values for each categorical column we can't use onehotencoder, we will have to manually assign the values to each type of categorical value.

```python
DF['cut']=np.where(DF['cut'] =='Ideal', 1.1, DF['cut'])
DF['cut']=np.where(DF['cut'] =='Premium', 1.2, DF['cut'])
DF['cut']=np.where(DF['cut'] =='Very Good', 1.3, DF['cut']) # Assigning values to cut feature
DF['cut']=np.where(DF['cut'] =='Good', 1.4, DF['cut'])
DF['cut']=np.where(DF['cut'] =='Fair', 1.5, DF['cut'])
```

```python
DF['color']=np.where(DF['color'] =='D', 2.1, DF['color'])
DF['color']=np.where(DF['color'] =='E', 2.2, DF['color'])
DF['color']=np.where(DF['color'] =='F', 2.3, DF['color'])
DF['color']=np.where(DF['color'] =='G', 2.4, DF['color'])    # Assigning values to color feature
DF['color']=np.where(DF['color'] =='H', 2.5, DF['color'])
DF['color']=np.where(DF['color'] =='I', 2.6, DF['color'])
DF['color']=np.where(DF['color'] =='J', 2.7, DF['color'])
```

```python
DF['clarity']=np.where(DF['clarity'] =='IF', 3.1, DF['clarity'])
DF['clarity']=np.where(DF['clarity'] =='VVS1', 3.2, DF['clarity'])
DF['clarity']=np.where(DF['clarity'] =='VVS2', 3.3, DF['clarity'])
DF['clarity']=np.where(DF['clarity'] =='VS1', 3.4, DF['clarity'])
DF['clarity']=np.where(DF['clarity'] =='VS2', 3.5, DF['clarity'])     # Assingning values to clarity feature
DF['clarity']=np.where(DF['clarity'] =='SI1', 3.6, DF['clarity'])
DF['clarity']=np.where(DF['clarity'] =='SI2', 3.7, DF['clarity'])
DF['clarity']=np.where(DF['clarity'] =='I1', 3.8, DF['clarity'])
```

```python
DF['cut']=DF['cut'].astype(np.float64)
DF['color']=DF['color'].astype(np.float64)    # Finally changing the data types from object to float
DF['clarity']=DF['clarity'].astype(np.float64)
```

# Now, let's start by splitting the dataset into training and testing.
## For this dataset, we will be using the 75:25 ratio.

```python
#Assigning the input and output/ predictor variable()
#Input variable is always 2-D and output varaible is always 1-D

X= DF.drop('price',axis=1)  # Al the features except the price column will be given as input 
y= DF[['price']] # We will be predicting the price
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25 , random_state=1) # dividing the data into training and testing
```

```python
LR = LinearRegression()   #  using linear regresssion
LR.fit(X_train, y_train) # fitting the model
```

```python
y_pred = LR.predict(X_test)  # predicting the price on the basis of testing data after training the model on training dataset
y_pred[7]
```
