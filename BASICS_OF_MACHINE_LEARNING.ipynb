```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/ameenmanna8824/DATASETS/main/areavsprices.csv")
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Area'],df['Prices'])
```

```python
#divide into input and output
#X is always input
#input is 2D and output is 1D
#adding : in column part gives 2D array
#without : , array will be 1D
```

```python
x=df.iloc[0:6,0:1].values # .values convert data into arrays
```

```python
y=df.iloc[0:6,1].values
```

```python
#Training the model
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x,y) #.fit helps in mapping all the inputs and outputs
```

```python
y_pred= model.predict(x) # predicting the output on the basis of input given
y_pred
```

```python
model.predict([[1400]])
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_pred,y)*100
```

```python
#now compare the predicted output with the actual output 
# if there is huge difference between these values, it means that the data is non linear
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred,c='orange')
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
df
```

```python
df.info()
```

```python
X=df.iloc[:,2:4].values # taking input, it is a 2D array
Y=df.iloc[:,4].values # taking output, it is a 1 D array
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train= scaler.fit_transform(x_train)
x_test= scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)*100
```

```python
#now check for any random values
model.predict([[20,25000]])
```

```python

from selenium import webdriver
from selenium.webdriver.firefox.service import Service as FirefoxService
from webdriver_manager.firefox import GeckoDriverManager

from bs4 import BeautifulSoup
import time
import pandas as pd
import numpy as np


def ScrapComment(url):
    #option = webdriver.EdgeOptions()
    #option.add_argument("--headless")
    driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))
    driver.get(url)
    prev_h = 0
    while True:
        height = driver.execute_script("""
                function getActualHeight() {
                    return Math.max(
                        Math.max(document.body.scrollHeight, document.documentElement.scrollHeight),
                        Math.max(document.body.offsetHeight, document.documentElement.offsetHeight),
                        Math.max(document.body.clientHeight, document.documentElement.clientHeight)
                    );
                }
                return getActualHeight();
            """)
        driver.execute_script(f"window.scrollTo({prev_h},{prev_h + 200})")
        # fix the time sleep value according to your network connection
        time.sleep(1)
        prev_h +=100 
        if prev_h >= height:
            break
    time.sleep(5)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    driver.quit()
    title_text_div = soup.select_one('#container h1')
    title = title_text_div and title_text_div.text
    comment_name=soup.select("#header-author #author-text")
    x=[]
    for i in range(len(comment_name)):
        x.append(comment_name[i].get_text().replace("  ",''))
    username=[]
    for i in range(len(x)):
        username.append(x[i].replace('\n',''))


    #name_list=[y.text for y in comment_name]
    comment_div = soup.select("#content #content-text")
    comment_list =[x.text for x in comment_div]
    print("Title of the video is: ",title.split("\n"))
    #print("Name: ",y)
    #print("List of the comment:",comment_list)
    resultdict = dict(zip(username,comment_list))
    #print(resultdict)
    df=pd.DataFrame(username,columns={'Username'})
    df['Comment']=comment_list

    DF= df.style.set_properties(**{'text-align': 'left'})
 
    DF= DF.set_table_styles(
    [dict(selector = 'th', props=[('text-align', 'left')])])
    print(DF)


urls= input("Enter YouTube video URL: ")# please enter the url of video which has only few comment
ScrapComment(urls)

```
